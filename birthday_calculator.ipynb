```python
from datetime import datetime, timedelta

# Function to calculate the age based on birthdate and current date
def calculate_age(birthdate, current_date):
    # Convert birthdate string to datetime object
    birthdate = datetime.strptime(birthdate, "%m/%d/%y")
    
    # Calculate age in years
    age = (current_date - birthdate).days // 365
    
    # If age is greater than 2, return age in years, otherwise return age in days
    if age > 2:
        return f"{age} years old"
    else:
        return f"{(current_date - birthdate).days} days old"

# Function to calculate the number of days until the next birthday
def calculate_days_until_birthday(birthdate, current_date, name):
    # Convert birthdate string to datetime object
    birthdate = datetime.strptime(birthdate, "%m/%d/%y")
    
    # Adjust birthdate to current year
    adjusted_birthdate = birthdate.replace(year=current_date.year)
    
    # Check if birthday is today, tomorrow, or yesterday
    if adjusted_birthdate.date() == current_date.date():
        return f"{name}'s birthday is today!"
    
    tomorrow = current_date + timedelta(days=1)
    if adjusted_birthdate.date() == tomorrow.date():
        return f"{name}'s birthday is tomorrow!"
    
    yesterday = current_date - timedelta(days=1)
    if adjusted_birthdate.date() == yesterday.date():
        return f"{name}'s birthday was yesterday!"
    
    # Calculate days until the next birthday
    if adjusted_birthdate > current_date:
        days_until_birthday = (adjusted_birthdate - current_date).days
        return f"{name}'s birthday is in {days_until_birthday} days"
    
    adjusted_birthdate = birthdate.replace(year=current_date.year + 1)
    days_until_birthday = (adjusted_birthdate - current_date).days
    return f"{name}'s birthday is in {days_until_birthday} days"

# Main function to interact with the user
def main():
    while True:
        print("Birthday Calculator")
        name = input("Enter name: ")
        birthdate = input("Enter birthday (MM/DD/YY): ")
        
        current_date = datetime.now()
        
        birthdate_obj = datetime.strptime(birthdate, "%m/%d/%y")
        
        # Print the entered birthdate and current date
        print("\nBirthday:  " + birthdate_obj.strftime("%A, %B %d, %Y"))
        print("Today:     " + current_date.strftime("%A, %B %d, %Y"))
        
        # Calculate and print the age
        age = calculate_age(birthdate, current_date)
        print(f"{name} is {age}.")
        
        # Calculate and print days until the next birthday
        days_until_birthday = calculate_days_until_birthday(birthdate, current_date, name)
        print(days_until_birthday + "\n")
        
        # Ask the user if they want to continue or exit
        continue_input = input("Continue? (y/n): ")
        if continue_input.lower() != 'y':
            print("\nBye!")
            break

# Entry point of the program
if __name__ == "__main__":
    main()
```
